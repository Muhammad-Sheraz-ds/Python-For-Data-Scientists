### Questin 01
Scientific Data Analysis with NumPy You are a scientist conducting research on climate change. You have collected temperature data from weather stations around the world stored in a NumPy array. Your task is to analyze the data using NumPy to:
1. Calculate the average temperature for each month across all stations.
2. Identify the station with the highest and lowest average temperature.
3. Determine the month with the largest temperature variation.
4. Visualize the temperature data using line plots or heatmaps.



<b>Use the following data set

```python
import numpy as np

# Sample temperature data for each month (in Celsius)
temperature_data = np.array([
    # January
    [10, 12, 15, 8, 9],
    # February
    [11, 13, 14, 9, 10],
    # March
    [12, 14, 16, 10, 11],
    # April
    [14, 16, 18, 12, 13],
    # May
    [16, 18, 20, 14, 15],
    # June
    [18, 20, 22, 16, 17],
    # July
    [20, 22, 24, 18, 19],
    # August
    [22, 24, 26, 20, 21],
    # September
    [20, 22, 24, 18, 19],
    # October
    [18, 20, 22, 16, 17],
    # November
    [15, 17, 19, 13, 14],
    # December
    [12, 14, 16, 10, 11]
])

# Station names for reference
station_names = ['Station A', 'Station B', 'Station C', 'Station D', 'Station E']
```

## Calculate the average temperature for each month across all stations.

```python
np.mean(temperature_data,axis=1)
```

## Identify the station with the highest and lowest average temperature

```python
station_wise_avg = np.mean(temperature_data,axis=0)
station_wise_avg
```

### Station of Highest Temperature

```python
station_wise_avg.max()
```

```python
np.where(station_wise_avg == station_wise_avg.max())[0][0]
```

```python
station_names[np.where(station_wise_avg == station_wise_avg.max())[0][0]]
```

### Station of Lowest Temperature

```python
station_names[np.where(station_wise_avg == station_wise_avg.min())[0][0]]
```

## Determine the month with the largest temperature variation

```python
max_temp = np.max(temperature_data,axis=1)
min_temp = np.min(temperature_data,axis=1)
min_temp
```

```python
variation = max_temp - min_temp
variation
```

```python
variation.max()
```

```python
np.where(variation == variation.max())
```

## Visualize the temperature data using line plots or heatmaps.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))
plt.plot(station_names,np.mean(temperature_data,axis=0))
plt.xlabel('Names Of Stations')
plt.ylabel('Temperature')
plt.title('Average Temperature in Stations')
plt.show();
```

```python
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.heatmap(temperature_data)
plt.title('Heatmap of Temperature Data')
plt.xlabel('Weather Stations')
plt.ylabel('Months')
plt.show();
```

## Calculate the cumulative returns over the year.

```python
np.cumprod(1 + daily_returns) - 1
```
